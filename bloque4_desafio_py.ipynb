```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. DataFrame
datos = {
    'reactor': ['R1','R1','R1','R2','R2','R2','R3','R3'],
    'turno': ['manana','tarde','noche','manana','tarde','noche','manana','tarde'],
    'temperatura': [85, 92, 78, 95, 88, 91, 76, 83],
    'eficiencia': [91.2, 87.5, 94.1, 83.3, 89.7, 85.0, 96.4, 92.8],
    'incidentes': [0, 1, 0, 2, 0, 1, 0, 0]
}

df = pd.DataFrame(datos)

# 2. Nueva columna 'estado'
df['estado'] = df['temperatura'].apply(lambda t: 'critico' if t > 90 else 'normal')

# 3. Conteo de estados
conteo_estado = df['estado'].value_counts()
print(conteo_estado)

# 4. Gráfico de barras: eficiencia promedio por reactor
ef_prom = df.groupby('reactor')['eficiencia'].mean()

plt.figure(figsize=(7, 4))

# colores por reactor (en orden R1, R2, R3)
colores = ['purple', 'darkblue', 'indigo']

plt.bar(ef_prom.index, ef_prom.values, color=colores)

plt.title('Eficiencia promedio por reactor')
plt.xlabel('Reactor')
plt.ylabel('Eficiencia (%)')
plt.tight_layout()
plt.show()
```
